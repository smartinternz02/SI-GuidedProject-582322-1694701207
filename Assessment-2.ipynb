```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
df=sns.load_dataset('car_crashes')
```

```
df
```

```
plt.scatter(df['total'],df['not_distracted'])  #Non-distracted case increases with the increase in the total cases.
```

```
plt.scatter(df['total'],df['speeding']) #We can say that speeding may result in more variable result than the not distracted once. But stil increaes as a whole
```
